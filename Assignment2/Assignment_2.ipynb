$$\begin{align}

Proof

\\=\left(\begin{array}{cccc}{s_{1}\sigma_{1}^2} & {s_{2}\sigma_{2}^2} & {\cdots} & {s_{p}\sigma_{p}^2}\end{array}\right)\left(\begin{array}{c}{s_{1}} \\ {s_{2}} \\ {\vdots} \\ {s_{p}}\end{array}\right)
\\=s_{1}^2\sigma_{1}^2+s_{2}^2\sigma_{2}^2+\cdots+s_{p}^2\sigma_{p}^2

Show that the emperical variance of the inner products of the columns of $\mathbf{X}$ with a normalized vector $\mathbf{a}$,

$$\begin{align}
\frac{1}{N} \sum_{i=1}^{N}\left(\mathbf{a}^{T} \mathbf{x}_{i}\right)^{2}=\frac{1}{N} \mathbf{a}^{T} \mathbf{X} \mathbf{X}^{T} \mathbf{a},
\end{align}
$$

is maximized when $\mathbf{a}$ is set to the first column of $\mathbf{U}$, i.e. $\mathbf{a}=\mathbf{u}_1$ (note that $\| \mathbf{a} \| = 1$).<br/>
Proof<br/>
Assume $\mathbf{a} \in \mathbb{R}^N$ is a linear combination of columns of $\mathbf{U}$.<br/>

$$\begin{align}
&\Rightarrow \mathbf{a}=\mathbf{Uc} \space(\space \mathbf{c} \space is \space the\space coefficient) 
\\
&\Rightarrow \|\mathbf{a}\|=\|\mathbf{U} \mathbf{c}\|=\sqrt{\mathbf{c}^{T} \mathbf{U}^{T} \mathbf{U} \mathbf{c}}=\sqrt{\mathbf{c}^{T} \mathbf{c}}=\|\mathbf{c}\|=1
\end{align}
$$

Using the SVD of $\mathbf{X}$ :
$$\begin{align}
\frac{1}{N} \mathbf{a}^{T} \mathbf{X} \mathbf{X}^{T} \mathbf{a}&=\frac{1}{N}(\mathbf{Uc})^{T} \mathbf{X} \mathbf{X}^{T} (\mathbf{Uc})\\
&=\frac{1}{N} \mathbf{c}^{T}\mathbf{U}^{T}\mathbf{X} \mathbf{X}^{T}\mathbf{Uc}\\
&=\frac{1}{N} \mathbf{c}^{T}\mathbf{U}^{T} ［\mathbf{U \Sigma V^{T}}］［\mathbf{U \Sigma V^{T}}］^{T} \mathbf{Uc}\\
&=\frac{1}{N} \mathbf{c}^{T}\left(\mathbf{U}^{T} \mathbf{U}\right) \mathbf{\Sigma}\left(\mathbf{V}^{T} \mathbf{V}\right) \mathbf{\Sigma}\left(\mathbf{U}^{T} \mathbf{U}\right) \mathbf{c}\\
&=\frac{1}{N} \mathbf{c}^{T} \mathbf{\Sigma}^{2} \mathbf{c}\\
&=\frac{1}{N} \sum_{i=1}^{N} \sigma_{i}^{2} c_{i}^{2}
\end{align}$$

Since $\sigma_1\ge\dots \sigma_p>0 $ and $\Vert \mathbf{c}\Vert=1$
$$\begin{align}
\max _{\|\mathbf{c}\|=1} \frac{1}{N} \mathbf{c}^{T} \Sigma \mathbf{\Sigma}^{T} \mathbf{c}=\max _{\|\mathbf{c}\|=1} \frac{1}{N} \sum_{i=1}^{N} \sigma_{i}^{2} c_{i}^{2}=\frac{1}{N} \sigma_{1}^{2}
\end{align}$$
When $\mathbf{c}=\left(\begin{array}{llll}{1} & {0} & {\dots} & {0}\end{array}\right)^{T}$ and $\left(\begin{array}{cccc}{-1} & {0} & {\ldots} & {0}\end{array}\right)^{T}$, we can attain the maximum.<br/> 

$$
\Rightarrow \mathbf{a}=\mathbf{Uc}=\mathbf{U}(\pm 1\quad 0\quad \ldots\quad 0)^{T}=\pm \mathbf{u}_{1}
$$
Hence $\mathbf{a}$ is the first column of $\mathbf{U}$.

```python
from PIL import Image
import numpy as np
from os import listdir
import imageio
from matplotlib import pyplot as plt
transform_of_pictures = {}
U_show = {}

I = Image.open('./subset0/person01_01.png')
I_column = I.size[1]*I.size[0]


pictures_name = listdir('subset'+str(0))

length = len(pictures_name)
index = 0
transform_of_pictures_train = np.zeros((I_column,length))
truth_label_train = np.zeros((length,1))
for picture in pictures_name:
    truth_label_train[index,:] = int(picture[6:8])
    file_index = './subset0' +  '/' + picture
    J = Image.open(file_index)
    transform_of_pictures_train[:,index] = np.reshape(J,(1,I_column))
    index=index+1
```

```python
def PCA_SVD(T,K):
    mean_of_data = np.mean(T,axis=1)
    mean_of_data = np.reshape(mean_of_data,(I_column,1))
    centured_data = T -  mean_of_data
    u,sigma,v_transport = np.linalg.svd(centured_data)
    u = u[:,0:K]
    return(u)
```

```python
T = transform_of_pictures_train
K = 20
U = PCA_SVD(T,K)
fig,axs = plt.subplots(1,3)
fig.suptitle('display first three vectors as images')
axs[0].imshow(U[:,0].reshape(I.size[1],I.size[0]))
axs[1].imshow(U[:,1].reshape(I.size[1],I.size[0]))
axs[2].imshow(U[:,2].reshape(I.size[1],I.size[0]))
```

#### Task 2.2

```python
Index_file = [1,2,3,4]
truth_label_test = {}
transform_of_pictures_test = {}

for i in Index_file:
    pictures_name = listdir('subset'+str(i))
    num_pictures = len(pictures_name)
    transform_of_pictures_test[i] = np.zeros((2500,num_pictures))
    truth_label_test[i] = np.zeros(num_pictures)
    index = 0
    for picture in pictures_name:
        truth_label_test[i][index] = int(picture[6:8])
        file_index = './subset'+str(i) +'/' + picture
        J = Image.open(file_index)
        transform_of_pictures_test[i][:,index] = np.reshape(J,(1,I_column))
        index+=1
     


```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
def error_rate_knn(T,labels_T,S,labels_S,U_first_step,K):
    
    #U_train = PCA_SVD(T,K)
    
    PC = np.transpose(U_first_step[:,0:K])
    T_reduced =np.dot(PC,T)
    S_reduced =np.dot(PC,S)
    neigb = knn(n_neighbors=3)
    neigb.fit(np.transpose(T_reduced),labels_T.ravel())
    error_rate = 1-neigb.score(np.transpose(S_reduced),labels_S) 
    return error_rate
    
    
    
    
    
    
```

```python
error_rate_result = np.zeros((4,20))
for K in range(1,21):
    for j in range(1,5):
        Train =  transform_of_pictures_train
        lab_T = truth_label_train
        Test  = transform_of_pictures_test[j]
        lab_S = truth_label_test[j]
        error_rate_result[j-1,K-1] = error_rate_knn(Train,lab_T,Test,lab_S,U,K)
```

```python
x = np.arange(1,21,1)
y_1 = error_rate_result[0,:]
y_2 = error_rate_result[1,:]
y_3 = error_rate_result[2,:]
y_4 = error_rate_result[3,:]
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)
plt.plot(x,y_4)
plt.xlabel('The number of principal componenets K')
plt.ylabel('Error rate for each subset')
plt.grid()
plt.legend(['subset 1','subset 2','subset 3','subset 4'])
plt.title('Error rate of each subset for K from 1 to 20')

```

### Conclusion and Analysis

Our conclusion: PCA does not necessarily improve classification ability of the samples. In other words, it cannot necessarily reduce the error rate.<br/>
Our analysis: PCA is able to reduce the dimension of data and the principal components (PCs) are based on the variance of the features for all samples in dataset. However, the classification ability does not equal to this variance and it is dependent on how strong the features are to separate the samples according to labels. It can also be affected by the difference of features between different class.
