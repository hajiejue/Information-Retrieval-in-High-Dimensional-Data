### Task 2.3+2.4

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
d = range(1,1001)
angle_aver = np.zeros((1001,1))
n = range(100)
angle = np.zeros((100,100))
angle_min = np.zeros((100,1))
for k in d:

    x = np.random.uniform(low=-1,high=1,size=(100,k))
    
    for i in n:
        for j in n:
            if i == j:
                angle[i,j] = 1.6
            else:    
                angle[i,j] = np.arccos( np.dot(x[i],x[j]) / ( np.sqrt(np.dot(x[i],x[i])) * np.sqrt( np.dot(x[j],x[j]) ) ) )
        
        angle_min = np.min(angle[i,:]) 
    
    angle_aver[k] = np.average(angle_min)
    
plt.plot(angle_aver)
plt.xlim(-50,1110)
plt.ylim(-0.2,2.0)
plt.xlabel('Dimension d')
plt.ylabel('Average of Minimum Angles(\omega)')
plt.title('Aver. Angle vs. Dimension')
```

Explanation of results:<br/>
(a) With the increase of the dimension of data, the average minimum angle between two randomly sampled vectors close to $\frac{\pi}{2}$,which means that the two randomly sampled vectors are orthogonal to each other. <br/>
(b) The reason therefore is following: for two d-dimensional vectors $a_1,a_2\in [-1,1]^d$ , the expected value of
$$
\begin{align}
\frac{|a_1\cdot a_2|^2}{|a_1|^2|a_2|^2} = \frac{1}{d} = E[cos^2(\omega)]
\end{align}
$$
$$
\begin{align}
d\to \infty,\quad E[cos^2(\omega)]\to0
\end{align}
$$
$$
\begin{align}
cos(\omega)=0,\quad \omega=\frac{\pi}{2}
\end{align}
$$
Conclusion:<br/>
When the dimension $d$ increases, the average minimum angle between two vectors $\in[-1,1]^d$ becomes $\frac{\pi}{2}$.

So $\mathbb{E}(\mathbf{X})=\mathbf{\mu}$.
To prove $\mathbf{C}=\mathbf{C}^T$:
$$
\begin{align*}
\mathbf{C} &= \mathbb{E}((\mathbf{X}-\mathbf{\mu})(\mathbf{X}-\mathbf{\mu})^T) \\
&= \mathbb{E}((\mathbf{X}-\mathbf{\mu})(\mathbf{X}^T-\mathbf{\mu}^T)) \\
&= \mathbb{E}(\mathbf{XX^T}-\mathbf{X}\mathbf{\mu}^T-\mathbf{X}^T\mathbf{\mu}+\mathbf{\mu}\mathbf{\mu}^T) \\
&= \mathbb{E}(\mathbf{XX^T})- \mathbb{E}(\mathbf{X})\mathbf{\mu}^T - \mathbb{E}(\mathbf{X}^T)\mathbf{\mu} + \mathbf{\mu}\mathbf{\mu}^T\\
